```python
test_case = int(input())
for tc in range(test_case):
    total_layer = int(input())
    print(f"#{tc+1}")
    # 이때의 tc는 각 단계에서의 pascal의 layer 수
    pascal = []
    # 0 ~ total_layer-1번쨰 layer 만들기
    for layer in range(total_layer):
        if layer == 0:
            pascal.append([1])
            print(*pascal[layer])
        else:
            pascal.append([1])
            for idx in range(0,layer-1):
                pascal[layer].append(pascal[layer-1][idx] + pascal[layer-1][idx+1])
            pascal[layer].append(1)
            print(*pascal[layer])
```

tc와 total layer는 다른데, Test Case input에서 tc가 total_layer와 같은 값으로 적혀있어서 해당 부분을 적용해주지 못하여 error가 발생했다.
