# Part3 01. 모험가 길드

각 모험가는 공포도(fear) 를 가지고 있음

어떤 그룹의 인원 수 ≥ 그 그룹에 있는 모든 모험가의 공포도

최대 그룹 수를 만들어야 함

몇 명의 모험가는 마을에 그대로 남아있어도 됨

```python
# 이 문제의 핵심은 마을에 사람이 많이 남아있어도 되니,
# 단순히 그룹의 수만을 최대화 하는 것이 핵심이기에
# 최대한 한 그룹 안에 사람이 적어야 그룹의 수 자체는 커진다.
```

```python
# 잘못 적은 초기 코드
n = int(input())
fear_list = list(map(int, input().split()))
fear_list.sort()
group = 0

while True:
    group_size = fear_list[-1]
    be_a_group = group_size
    for _ in range(group_size):
        if len(fear_list) > 0:
            fear_list = fear_list[0:-1]
            be_a_group -= 1
    if be_a_group == 0:
        group += 1

    if len(fear_list) == 0:
        break

print(group)
```

```python
# 단순히 순차적으로 group을 만들면서
# 그룹 안의 원소의 수와 그룹 안에서 가장 큰 원소의 크기가 같아지는 순간이
# 하나의 그룹이 완성되는 순간이기에, 그룹 안의 원소의 수는 초기화하여 다시 반복

n = int(input())
fear_list = list(map(int, input().split()))

fear_list.sort()

group = 0
present_group_elem_count = 0

for fear in fear_list:
    present_group_elem_count += 1

    if present_group_elem_count == fear:
        group += 1
        present_group_elem_count = 0

print(group)
```

# Part 3 02. 곱하기 혹은 더하기

```python
number = input()

result = 0

for num in number:
    result = max(result * int(num), result + int(num))

print(result)
```

# Part 3 03. 문자열 뒤집기

```python
# 초기 코드
zero_one_str = input()
before = int(zero_one_str[0])
change = 0
if len(zero_one_str) > 1:
    for num in zero_one_str[1:]:
        if before != int(num):
            change += 1
        before = int(num)

if change % 2 == 0:
    print(int(change / 2))
else:
    print(int((change + 1) / 2))
```

```python
# 개선 코드 : 보통은 i를 바탕으로 i-1 번째 숫자와 비교함
# Just 적게 나온 숫자의 수만큼 변환을 주면 되기에 둘 중 더 적은 등장 횟수만큼만 반환

s = input()

count0 = 0
count1 = 0

if s[0] == '1':
    count0 += 1
else:
    count1 += 1

for i in range(1, len(s)):
    if s[i] != s[i-1]:
        if s[i] == '1':
            count0 += 1
        else:
            count1 += 1

print(min(count0, count1))
```

# Part 3 04. 만들 수 없는 금액

```python
# 초기 코드 : O(2^N)의 시간 복잡도
# 지수 함수의 시간 복잡도는 위험하다.

num_of_coin = int(input())
coin_list = list(map(int, input().split()))

maximum = sum(coin_list)
all_number = {0}

for num in coin_list:
    temp_list = []
    for set_num in all_number:
        temp_list.append(set_num + num)
    all_number = all_number | set(temp_list)

for i in range(1, maximum+1):
    if i not in all_number:
        print(i)
        break

    if (i == maximum) and i in all_number:
        print(maximum+1)
```

```python
n = int(input())
coins = list(map(int, input().split()))

coins.sort()

target = 1 # 기본적으로 없는 1을 하나 더 두어,
# 현재는 만들어 낼 수 없는 최소값

for coin in coins:
    if coin > target:
        break
    target += coin

print(target)
```

# Part 3 05. 볼링공 고르기

```python
N, M = map(int, input().split())
num_list = list(map(int, input().split()))

num_list.sort()

result = 0
remain = N

for num in range(1, M+1):
    remain -= num_list.count(num)
    result += remain * num_list.count(num)

print(result)
```

```python
# 개선된 코드 :
# Count는 1번에 O(N)만큼의 연산을 하기에 여러번 반복하는 것은 좋지 않다.

N, M = map(int, input().split())
num_list = list(map(int, input().split()))

num_list.sort()
list_count = [0] * M

result = 0
remain = N

for num in num_list:
    list_count[num-1] += 1

for num in range(0, M):
    remain -= list_count[num]
    result += remain * list_count[num]

print(result)
```

# Part 3 06. 무지의 먹방 라이브

```python
import heapq

def solution(food_times, k):

    if sum(food_times) <= k:
        return -1

    n = len(food_times)
    q = []

    for i in range(len(food_times)):
        # 음식 섭취 시간, 음식 index
        heapq.heappush( q ,(food_times[i], i+1))

    # 지금까지 지난 총 시간
    total_time = 0

    # 바로 직전에 다같이 차감된 개별 시간
    sum_previous_time = 0

    for i in range(len(food_times)):
        now = heapq.heappop(q)
        if (now[0] - sum_previous_time) * n <= k - total_time:
            total_time += n * (now[0] - sum_previous_time)
            n = n - 1
            sum_previous_time += now[0] - sum_previous_time
        else:
            heapq.heappush(q, now)
            break

    remain = sorted(q, key = lambda x : x[1])
    return remain[(k - total_time) % n][1]

print(solution([4, 1, 1, 5], 4) )
```
