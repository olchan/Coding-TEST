```python
"내부 값을 바꾸는 것은 같은 객체를 수정하는 것이고"
"a[0] = 3은 객체를 바꾸는 게 아니라, 그 위치의 참조를 다른 객체로 교체하는 것이다."

a = [[0]]*3
print(a)
a[0] = [3] # 가장 중요한 것은 객체를 직접 수정한 것인지 (mutation)
# 아니면 참조를 다른 객체로 교체한 것인지 (re-binding) 이다.
print(a)
```

> ### 우선 순위 큐 : heapq

- 우선 순위 큐는 Heap이랑 거의 같은 개념이다. (우선 순위 큐를 구현하는 대표적인 자료구조가 queue인 것)

```python
# Min HEAP
import heapq

pq = []
heapq.heappush(pq, 5)
heapq.heappush(pq, 2)
heapq.heappush(pq, 8)

print(heapq.heappop(pq))
```

```python
# Max HEAP : 내림차순 정렬
heapq.heappush(pq, -5)
heapq.heappush(pq, -2)
heapq.heappush(pq, -8)
max_num = -heapq.heappop(pq)
max_num
```

```python
# 주로, 우선순위 + 값 같이 저장됨
# heapq.heappush(pq, (priority, data))
pq = []
heapq.heappush(pq, (3, "A"))
heapq.heappush(pq, (1, "B"))
heapq.heappush(pq, (2, "C"))
```

```python
# 한번에 heap 구조로 바꿔주기

h = [1,2,6,3,7,8,5,9,0]
heapq.heapify(h)
print(h)
```

> 아래처럼 heap을 사용한다고 해서 내림차순 정렬이 되는 것은 아니다. Heap은 단지 부모 <= 자식일 뿐이지, 전체 정렬이 아니기 때문이다. 제대로 된 정렬을 구현하기 위해서는 heappop()을 반복해서 구현해야한다.

```python
# 잘못된 구현
import heapq

def heapsort(iterable):
  iterable = list(map(lambda x: -x, iterable))
  heapq.heapify(iterable)
  iterable = list(map(lambda x: -x, iterable))
  return iterable

result = heapsort([5,3,2,7,4,8])
print(result)
```

```python
# 제대로 된 구현
import heapq

def heapsort(iterable):
  h = []
  result = []
  for i in iterable:
    heapq.heappush(h, -i)
  for k in range(len(h)):
    result.append(-heapq.heappop(h))
  return result

result = heapsort([5,3,2,7,4,8])
print(result)
```

> ### 정렬된 배열에서 이진 탐색으로 삽입 위치를 찾아주는 라이브러리 : bisect
- bisect_left() & bisect_right() 함수가 가장 중요하게 사용된다. -> 시간복잡도 : O(logN)

image.png

```python
import bisect

a = [1,2,4,4,8]
x = 4

print(bisect.bisect_left(a, x))
print(bisect.bisect_right(a, x))
```

정렬된 배열에서는 x의 개수를 셀 때, 낮은 시간 복잡도로 개수를 계산 가능하다.
- count를 여러번 하는 경우에는 한번 정렬(n log n)해두고, 여러번 bisect로 count (log n)가 더 효율적이지만, count를 많이 하지 않는다면, 정렬 없이 count가 더 낫다.

```python
left = bisect.bisect_left(a, x)
right = bisect.bisect_right(a, x)
count = right - left
count
```

```python
# 정렬된 list에서 left value <= x <= right value 인 원소의 개수를 o(logN)으로 빠르게 계산
from bisect import bisect_left, bisect_right
def count_by_range(arr, left_bisect, right_bisect):
  right_index = bisect_right(arr, right_bisect)
  left_index = bisect_left(arr, left_bisect)
  return right_index - left_index

a = [1,2,4,4,5,5,5,6,6,7,7,7,7,8]

print(count_by_range(a, 4, 4))
print(count_by_range(a, 3, 5))
```

> Itertools

```python
# Permutation:  iterable 객체에서 r개의 data를 뽑아
# 일렬로 나열하는 모든 경우를 계산
from itertools import permutations

data = ["A", "B", "C"]
result = list(permutations(data, 2))
print(result)


# Combination
from itertools import combinations
data = ["A", "B", "C"]
result = list(combinations(data, 2))
print(result)

# product : 중복을 허용하여 repeat 만큼의 원소 뽑기
from itertools import product
data = ["A", "B", "C"]
result = list(product(data, repeat = 2))
print(result)
```

>  2차원 List (행렬)을 90도 회전하는 method

```python
def rotate_a_matrix_by_90_degree(a):
  row_length = len(a)
  column_length = len(a[0])

  res = [[0] * row_length for _ in range(column_length)]
  for r in range(row_length):
    for c in range(column_length):
      res[c][row_length - r - 1] = a[r][c]
  return res
```
