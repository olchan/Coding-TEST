```python
숫자 N은 아래와 같다.

N=2^a x 3^b x 5^c x 7^d x 11^e

N이 주어질 때 a, b, c, d, e 를 출력하라.


[제약 사항]

N은 2 이상 10,000,000 이하이다.


[입력]

가장 첫 줄에는 테스트 케이스의 개수 T가 주어지고, 그 아래로 각 테스트 케이스가 주어진다.

각 테스트 케이스의 첫 번째 줄에 N 이 주어진다.


[출력]

출력의 각 줄은 '#t'로 시작하고, 공백을 한 칸 둔 다음 정답을 출력한다.

(t는 테스트 케이스의 번호를 의미하며 1부터 시작한다.)
```

```python
# 입력
10
6791400
1646400
1425600
8575
185625
6480
1185408
6561
25
330750

# 출력
#1 3 2 2 3 1
#2 6 1 2 3 0
#3 6 4 2 0 1
#4 0 0 2 3 0
#5 0 3 4 0 1
#6 4 4 1 0 0
#7 7 3 0 3 0
#8 0 8 0 0 0
#9 0 0 2 0 0
#10 1 3 3 2 0
```

```python
test_num = int(input())

for i in range(1, test_num+1):
    list_jaegob = [0,0,0,0,0]
    list_jagob_num = [2,3,5,7,11]
    input_number = int(input())
    for index in range(len(list_jagob_num)):
        copy_input_number = input_number
        repeat = True
        while(repeat):
            if (copy_input_number % list_jagob_num[index] == 0):
                copy_input_number = copy_input_number // list_jagob_num[index]
                list_jaegob[index] += 1
            else:
                repeat = False
    print(f"#{i}", *list_jaegob ,sep = ' ')
```

> 개선된 Code

```python
test_num = int(input())

for i in range(1, test_num+1):
    list_jaegob = [0]*5 # [0,0,0,0,0]
    list_jagob_num = [2,3,5,7,11]
    input_number = int(input())
    for index in range(len(list_jagob_num)):
        while(input_number % list_jagob_num[index] == 0):
          # repeat = True 대신 조건 자체를 while에 넣기
          # 어차피 소인수 분해이니까, 나눈 값으로 값 udpate
            input_number = input_number // list_jagob_num[index]
            list_jaegob[index] += 1

    print(f"#{i}", *list_jaegob ,sep = ' ')
```
