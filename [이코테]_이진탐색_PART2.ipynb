### 순차 탐색
- 리스트 안에 있는 특정한 data를 찾기 위해 앞에서부터 data를 하나씩 차례대로 확인하는 방법
- 정렬되지 않은 리스트에서 data를 찾아야 할 때 사용

```python
def sequential_search(n, target, array):
  for i in range(n):
    if array[i] == target:
      return i
```

### 이진 탐색

- 정렬된 배열에서만 사용 가능한 탐색 알고리즘
- start, end, middle의 3개의 변수를 사용해서 찾고자 하는 data와 middle 위치에 있는 data의 크기를 비교해서 탐색 범위를 좁혀나간다.

- 시간 복잡도 : 1번의 탐색마다 확인해야하는 data 수가 1/2로 줄어든다는 점에서 O(logN)의 시간 복잡도를 가진다.


```python
# 재귀함수로 이진 탐색 구현하기

def binary_search(array, target, start, end):
    if start > end:
        print("원소가 존재하지 않습니다.")
        return

    mid = (start + end) // 2 # 정수 index 반환

    if array[mid] == target:
        print (mid + 1)
        return

    elif array[mid] > target:
        end = mid - 1
        return binary_search(array, target, start, end)
        # return 값이 있다면, 무조건 위와 같이 해야한다.

    else:
        start = mid + 1
        return binary_search(array, target, start, end)


arr = [2,4,6,8,9, 11, 13, 15]
binary_search(arr, 13, 0, len(arr)-1)
```

```python
# 반복문으로 이진 탐색 구현하기

def binary_search(array, target, start, end):
  while start <= end:
    mid = (start + end) // 2

    if array[mid] == target:
      return mid

    elif array[mid] > target:
        end = mid - 1

    else:
        start = mid + 1

  return None

arr = [2,4,6,8,9, 11, 13, 15]
result = binary_search(arr, 13, 0, len(arr)-1)
if result == None:
  print("원소가 존재하지 않습니다.")
else:
  print(result)
```

```python
# 집합 자료형을 사용했을 때
n = int(input())
part_num = list(map(int, input().split()))
m = int(input())
want_to_found = list(map(int, input().split()))

set_part_num = set(part_num)
for item in want_to_found:
    if item in set_part_num:
        print("yes", end = " ")
    else:
        print("no", end = " ")
```

```python
# 계수 정렬을 사용했을 때
n = int(input())
part_num = list(map(int, input().split()))
m = int(input())
want_to_found = list(map(int, input().split()))

minimum_part = min(part_num)
maximum_part = max(part_num)

count_list = [0] * (maximum_part - minimum_part +1)
for item in part_num:
    count_list[item - minimum_part] += 1

for item in want_to_found:
    if count_list[item - minimum_part] > 0:
        print("yes", end = " ")
    else:
        print("no", end = " ")
```

```python
# 이진 탐색을 사용했을 때
n = int(input())
part_num = list(map(int, input().split()))
part_num.sort()
m = int(input())
want_to_found = list(map(int, input().split()))

def binary_search(array, target, start, end):
    while start <= end:
        mid = (start + end) // 2
        if array[mid] == target:
            return True
        elif array[mid] > target:
            end = mid - 1
        else:
            start = mid + 1
    return False

for item in want_to_found:
    if binary_search(part_num, item, 0, len(part_num) - 1):
        print("yes", end = " ")
    else:
        print("no", end=" ")
```

> ### 실전 문제 3. 떡볶이 떡 만들기

```python
# 경계값 찾기 문제 : 이진 탐색!!
"""
길이가 긴 h부터 탐색하면서, 총합 길이 m을 만족시키는 h를 찾자.
-> 후보를 하나씩 다 보는 방식은 조건을 만족하는 h가 작아질수록 비용이 너무 커짐

-> “어떤 h가 가능하면 그보다 작은 h는 전부 가능하고,
어떤 h가 불가능하면 그보다 큰 h는 전부 불가능하네?”

-> "가능 가능 가능 가능 가능 불가능 불가능 불가능" 과 같은 구조이다보니,
경계값을 찾는 문제임을 알 수 있고, 이를 통해 이진 탐색 구조임을 알 수 있다.
"""
```

> 시간 복잡도 : O(N log K)

```python
"""
바깥: 이진 탐색 → log K 번
안쪽: 매번 전체 순회 → O(N)
"""
# 왜 end를 최종적으로 출력하는가?
# mid에서 조건을 만족하면 start가 mid+1로 이동하고,
# 이후 탐색에서 end가 줄어들면서
# 최종적으로 end가 그 mid 값에 위치하게 된다.
```

```python
n, m = map(int, input().split())
list_num = list(map(int, input().split()))

start = 0
end = max(list_num) #O(N)

# “더 큰 길이에서도 m을 만들 수 있으니까 멈추면 안 된다”
while start <= end:
    remain_len = 0
    mid = (start + end) // 2
    for x in list_num:
        if x > mid:
            remain_len += (x - mid)
    # if remain_len == m:
    #     break
    if remain_len >= m:
        start = mid + 1
        # start와 end가 같았다면, end가 mid 이므로 end 출력
    else:
        end = mid - 1
        # remain_len 이 m을 만족시키지 못했기에, end를 더 줄였고,
        # 그러다가 아예 만족을 못 시키면, -1을 반환

print(end)
```

```python
# 조금 더 직관적인 code
# 조건을 만족하는 mid를 계속 result에 update하면서,
# 이를 만족하는 h를 끝까지 탐색

n, m = map(int, input().split())
array = list(map(int, input().split()))

start = 0
end = max(array) # O(N)

result = 0

while start <= end:
    total = 0
    mid = (start + end) // 2
    for x in array:
        if x > mid:
            total += x - mid

    if total < m:
        end = mid -1
    elif total >= m:
        result = mid
        start = mid + 1

print(result)
```
